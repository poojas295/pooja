```python

tran_df = pd.read_csv(io.BytesIO(blob_tran_data))
tran_df
```

```python
cus_df = pd.read_csv(io.BytesIO(blob_cus_data))
cus_df
```

```python
ord_df = pd.read_csv(io.BytesIO(blob_ord_data))
ord_df
```

```python
df1 = ord_df.merge(tran_df, on= 'order_id', how= 'inner')
df1.head()
```

```python
df2 = df1.merge(cus_df, how= 'inner', on= 'customer_id')
df2
```

```python
# group by customer ID and calculate average purchase amount
average_purchase = df2.groupby(['customer_id']).mean()
average_purchase
```

```python
# filter for customers with average purchase amount > $400
high_spenders = average_purchase[average_purchase['sales'] > 400]

high_spenders
```

```python
# output list of high spending customers
print(high_spenders[['sales', 'profit']])
```
